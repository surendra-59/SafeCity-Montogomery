```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Dataset/311_requests_full.csv', low_memory=False)

# Basic stats and structure
print("--- Dataset Shape ---")
print(df.shape)

print("\n--- Data Types & Nulls ---")
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': df.isnull().sum(),
    'Unique Values': df.nunique()
})
print(summary)

print("\n--- First 5 Rows ---")
display(df.head())
```

```python
# Inspecting distinct values for the requested columns
cols_to_check = ['Department', 'Request_Type', 'Origin']

for col in cols_to_check:
    print(f"--- Unique Values in {col} ({df[col].nunique()}) ---")
    # Showing top 15 by frequency to see the "big players"
    print(df[col].value_counts().head(15))
    print("\n")

# Quick check on the 'Status' as well since it only has 4 unique values
print("--- Status Distribution ---")
print(df['Status'].value_counts())
```

```python
# Load the Code Violations dataset
df_violations = pd.read_csv('Dataset/montgomery_code_violations_full.csv', low_memory=False)

# 1. Basic Health Check
print("--- Violations Dataset Shape ---")
print(df_violations.shape)

print("\n--- Column Overview & Nulls ---")
summary_v = pd.DataFrame({
    'Data Type': df_violations.dtypes,
    'Missing Values': df_violations.isnull().sum(),
    'Unique Values': df_violations.nunique()
})
print(summary_v)

# 2. Categorical Deep Dive
# Identify the top violation types and their status
print("\n--- Top 10 Violation Types ---")
if 'Violation_Type' in df_violations.columns: # Adjust name if different
    print(df_violations['Violation_Type'].value_counts().head(10))

print("\n--- Violation Status ---")
if 'Status' in df_violations.columns:
    print(df_violations['Status'].value_counts())

# 3. Preview
display(df_violations.head())
```

```python
# 1. Distinct Case Types
print("--- Case Type Distribution ---")
print(df_violations['CaseType'].value_counts(dropna=False))

# 2. Case Status (Since there are only 3)
print("\n--- Case Status Breakdown ---")
print(df_violations['CaseStatus'].value_counts())

# 3. Lien Status (Checking the 65k nulls)
print("\n--- Lien Status ---")
print(df_violations['LienStatus'].value_counts(dropna=False))
```

```python
# Load the Code Violations dataset
df_violations = pd.read_csv('Dataset/Weather_Sirens.csv', low_memory=False)

# 1. Basic Health Check
print("--- Weather_Sirens Dataset Shape ---")
print(df_violations.shape)

print("\n--- Column Overview & Nulls ---")
summary_v = pd.DataFrame({
    'Data Type': df_violations.dtypes,
    'Missing Values': df_violations.isnull().sum(),
    'Unique Values': df_violations.nunique()
})
print(summary_v)

print("\n--- Status ---")
if 'Status' in df_violations.columns:
    print(df_violations['Status'].value_counts())

# 3. Preview
display(df_violations.head())
```

```python
# 1. Clean column names (strip whitespace)
df.columns = df.columns.str.strip()
df_violations.columns = df_violations.columns.str.strip()

# 2. Print columns just to be 100% sure of the spelling
print("311 Columns:", df.columns.tolist())
print("Violation Columns:", df_violations.columns.tolist())

# 3. Try conversion again with flexible column selection
# We'll use a small loop to handle the date conversion safely
for col in ['Create_Date', 'CaseDate', 'Close_Date']:
    if col in df.columns:
        df[f'{col}_DT'] = pd.to_datetime(df[col], unit='ms', errors='coerce')
    if col in df_violations.columns:
        # Violations usually come in as strings, so no 'unit' needed
        df_violations[f'{col}_DT'] = pd.to_datetime(df_violations[col], errors='coerce')

# 4. Verification
print("\n--- Date Conversion Check ---")
print(df[['Create_Date', 'Create_Date_DT']].head(2) if 'Create_Date' in df.columns else "311 Date Missing")
print(df_violations[['CaseDate', 'CaseDate_DT']].head(2) if 'CaseDate' in df_violations.columns else "Violation Date Missing")
```

```python
# 1. Check the very end of the Violations columns
print("Last 10 columns of Violations:")
print(df_violations.columns[-10:].tolist())

# 2. Check 311 Date Range
print("\n--- 311 Date Range ---")
print(f"Start: {df['Create_Date_DT'].min()}")
print(f"End:   {df['Create_Date_DT'].max()}")

# 3. Search for ANY date-like column in the Violations dataset
date_cols = [c for c in df_violations.columns if 'date' in c.lower() or 'year' in c.lower()]
print("\nPotential Date Columns in Violations:", date_cols)
```

```python
# 1. Look for a sample of the data to see what's actually inside
print("--- Data Sample (Transposed for readability) ---")
display(df_violations.head(3).T)

# 2. Search for the 'CaseType' or 'NUISANCE' values specifically
# This checks if the data is there but the headers are renamed
has_nuisance = df_violations.stack().str.contains('NUISANCE', na=False).any()
print(f"\nDoes the word 'NUISANCE' exist in the data? {has_nuisance}")

# 3. Check 311 Volume by Year
print("\n--- 311 Requests by Year ---")
print(df['Create_Date_DT'].dt.year.value_counts().sort_index())
```
